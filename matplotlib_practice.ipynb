## Create a Line Chart using the MATLAB method

```python
saving=[0,100,150,200,300,500,300,350,500,600,750,1000]
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(months,saving)
plt.xlabel("Months")
plt.ylabel("Savings ($)")
plt.title("Savings Account Results")
#plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with plt.plot() --- 'plt.plot()' is the MATLAB method
plt.plot(x_axis, y_axis)
```

## Create a Line Chart using the Object-Oriented Interface Method

```python
# Create the plot with ax.plt() --- 'ax.plot()' is the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

## Create a Pie Chart using the MATLAB Method 

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

**Add % to the wedges**

```python
plt.pie(y_axis,labels=x_axis, autopct="%1.1f%%")
plt.show()
```

**Explode the largest percentage** 
- Explode values are 12, one for each value in the pie Chart. 
The value specified to explode a wedge is he fraction of the radius from the center of the pie for each wedge.
- To pop out the seventh month, July, we use a decimal-point value of 0.2.
- To add the percentage of each wedge on the pie chart, we use the autopct parameter and provide the format of one decimal place using .1f%.
- The % before and after the .1f% formats the number as a percentage.


```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,labels=x_axis, autopct="%1.1f%%", explode=explode_values)
plt.show()
```

**Change the color of each pie wedge**
- It is a best practice to choose lighter, brighter, and analogous colors over darker tones.
- We need to assign a "colors" variable to an array with the names of our 12 colors.
- We'll add the "colors" parameter in the pie() function with our other parameters
- List of colors https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
        labels=x_axis, 
        autopct="%1.1f%%", 
        explode=explode_values,
        colors = colors)
plt.show()
```

## Create a Pie Chart using the Object-Oriented Method 

```python
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

**Skill Drill**
- Using the object-oriented approach, make the following changes to the pie chart:
- Add a percentage to one decimal place to each wedge of the pie.
- Increase the figure size to 8x8.
- Explode the two highest percentage months.
- Add a shadow.
- Add a start angle so that January is at the top.
- Reverse the order so that the month order is in a clockwise direction.
- Add new colors of your own choosing or use the colors from the previous pie chart.

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
explode_values=(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
ax.pie(y_axis,
      labels=x_axis,
      autopct="%1.1f%%",
      shadow=True,
      startangle=90,
      counterclock=False,
      colors=colors,
      explode=explode_values
      )

plt.show()
```

## Change the Major Ticks

**Using MATLAB Approach**

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

**Using Object-Oriented Approach**

## Add Minor Ticks

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
