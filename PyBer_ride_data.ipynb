# Plot a Pandas DataFrame & Series

```python
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
import statistics
```

```python
path=("Resources/PyBer_ride_data.csv")
pyber_ride_df=pd.read_csv(path)
```

```python
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

**Bar Chart**

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])

pyber_ride_df.plot.bar(x="Month", 
                     y="Avg. Fare ($USD)",
                     color= "cyan",
                     yerr=stdev, 
                      capsize=3 )
plt.yticks(np.arange(0, 51, step=5.0))
plt.show()
```
